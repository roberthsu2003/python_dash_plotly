```python
import pandas as pd
#æ¥äººå£
proverty_data = pd.read_csv('PovStatsData.csv')
```

```python
regions = ['East Asia & Pacific', 'Europe & Central Asia',
           'Fragile and conflict affected situations', 'High income',
           'IDA countries classified as fragile situations',
           'IDA total', 'Latin America & Caribbean', 'Low & middle income',
           'Low income', 'Lower middle income', 'Middle East & North Africa',
           'Middle income', 'South Asia', 'Sub-Saharan Africa',
           'Upper middle income', 'World','West Bank and Gaza']
```

```python
population_df = proverty_data[~proverty_data['Country Name'].isin(regions) & (proverty_data['Indicator Name']=='Population, total')]
population_df
```

```python
year = '2010'
year_df = population_df[['Country Name', year]].sort_values(year,ascending=False)[:20]
year_df
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_bar(
    x=year_df['Country Name'],
    y=year_df[year]
)
fig.update_layout(title_text=f'æå¤äººå£ç20åå®¶-{year}')
fig.show()
```
