```python
import pandas as pd
sales = pd.read_csv('train2.csv')
sales['Order Date'] = pd.to_datetime(sales['Order Date'],format='%d/%m/%Y')
sales['Year'] = sales['Order Date'].dt.year
sales.head()
```

```python
sales['部分'].unique()
```

```python
sales1 = sales.groupby(['Year', '子類別', '部分'])['Sales'].sum().reset_index()
sales1
```

```python
sales2 = sales1[(sales1['Year'] == 2018)&(sales1['部分'] == '在家辦公')].sort_values(by='Sales',ascending=False).nlargest(5,columns=['Sales'])
sales2
```
