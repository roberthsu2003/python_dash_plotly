```python
import pandas as pd
sales = pd.read_csv('train2.csv')
sales['Order Date'] = pd.to_datetime(sales['Order Date'],format='%d/%m/%Y')
sales['Year'] = sales['Order Date'].dt.year
sales['Month'] = sales['Order Date'].dt.month
sales.head()
```

```python
sales['部分'].unique()
```

```python
sales1 = sales.groupby(['Year', '子類別', '部分'])['Sales'].sum().reset_index()
sales1
```

```python
sales2=sales1[(sales1['Year'] == 2018) & (sales1['部分'] == '公司的')].sort_values(by='Sales',ascending=False).nlargest(5,columns=['Sales'])
sales2
```

```python
sales3 = sales.groupby(['Year', '地區', '部分'])['Sales'].sum().reset_index()
sales3
```

```python
sales4 = sales3[(sales3['Year'] == 2018)&(sales3['部分'] == '在家辦公')].sort_values(by='Sales',ascending=False)
sales4
```

```python
sales['類別'].unique()
```

```python
sales5 = sales.groupby(['Year', '類別', '部分'])['Sales'].sum().reset_index()
sales5.head()
```

```python
sales_furniture = sales5[(sales5['Year']==2018) & (sales5['部分']=='公司的') & (sales5['類別']=='家具')]['Sales']
sales_furniture
```

```python
sales_office = sales5[(sales5['Year']==2018) & (sales5['部分']=='公司的') & (sales5['類別']=='辦公用品')]['Sales']
sales_office
```

```python
sales_technology = sales5[(sales5['Year']==2018) & (sales5['部分']=='公司的') & (sales5['類別']=='技術')]['Sales']
sales_technology
```

```python
sales6 = sales.groupby(['Year', 'Month', '部分'])['Sales'].sum().reset_index()
sales6
```

```python
sales7 = sales6[(sales6['Year']==2017) & (sales6['部分']=='公司的')].sort_values(by='Month')
sales7.head()
```

```python
sales8 = sales.groupby(['Year'])['Sales'].sum().reset_index()
sales8.head()
```

```python
current_year = sales8[sales8['Year']==2015]['Sales'].sum()
current_year
```

```python
sales9 = sales.groupby(['Year'])['Sales'].sum().reset_index()
sales9
```

```python
sales9['py'] = sales9['Sales'].shift(1)
sales9
```

```python
previous_year = sales9[sales9['Year'] == 2018]['py'].sum()
previous_year
```

```python
sales10 = sales.groupby(['Year'])['Sales'].sum().reset_index()
sales10['yoy growth'] = sales10['Sales'].pct_change()
sales10['yoy growth'] = sales10['yoy growth'] * 100
sales10
```

```python
yoy_growth = sales10[sales10['Year']==2018]['yoy growth'].sum()
yoy_growth
```
